```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from math import pi
excel_df = pd.read_excel("Najgorsze.xlsx")
```

```python
# Set data
df = pd.DataFrame({
'Wojewdztwo': excel_df['Wojewdztwo'],
'Przygotowanie szpitali': excel_df['Przygotowanie'],
'Rzeczywista sytuacja w szpitalach ': excel_df['Rzeczywisto '],
'Testowanie': excel_df['Testowanie'],
'Zakoczenie choroby': excel_df['Skutek choroby'],
'Procentowa liczba chorych': excel_df['% chorych']
})
 
 
# ------- PART 1: Create background
categories=list(df)[1:]
N = len(categories)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
 
ax = plt.subplot(111, polar=True)
 
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
ax.patch.set_alpha(0)
plt.xticks(angles[:-1], categories)
 
ax.set_rlabel_position(1)
plt.yticks([10,20,30,40,50,60,70,80,90], ["10","20","30","40","50","60","70","80","90"], color="dimgrey", size=8)
plt.ylim(0,100)
plt.title('Najgorzej radzce sobie wojewdztwa', size=15)
 
 
# ------- PART 2: Add plots
# Ind1
values=df.loc[0].drop('Wojewdztwo').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1,color='crimson', linestyle='solid', label=df['Wojewdztwo'][0])
ax.fill(angles, values, 'crimson', alpha=0.2)
 
# Ind2
values=df.loc[1].drop('Wojewdztwo').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1,color='orangered', linestyle='solid', label=df['Wojewdztwo'][1])
ax.fill(angles, values, 'orangered', alpha=0.2)

# Ind3
values=df.loc[2].drop('Wojewdztwo').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1,color='orange', linestyle='solid', label=df['Wojewdztwo'][2])
ax.fill(angles, values, 'orange', alpha=0.2)
```
