```python
import numpy as np
import matplotlib.pyplot as plt
```

## Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

```python
x = np.linspace(-3, 3, num=1000)
y = x**3 + 1
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.ylabel('x^3 + 1')
plt.xlabel('x')
plt.title("y = x^3 + 1", loc = "center")
plt.show()
```

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

```python
plt.plot(x, y, label = "x^3 + 1")
plt.plot(x, y + 10, label = "x^3 + 11", linewidth = 3)
plt.legend(loc = "lower right")
plt.show()
```

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
fig, axs = plt.subplots(2)

axs[0].plot(x, y)
axs[1].plot(x, y + 10, color = "#FB871A", linewidth = 3)
```

## Zadanie 2

```python
labels = ['Blue', 'Red', 'Green', 'Purple']
sizes = [15, 30, 45, 10]
```

a) Narysować wykres słupkowy na podanych danych.

```python
plt.bar(labels, sizes)
plt.show()
```

b) Znaleźć sposób (w internecie) na narysowanie wykresu kołowego na podanych danych.

```python
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, colors = labels)
ax1.axis('equal')
plt.show()
```

## Zadanie 3

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej. Jest przynajmniej kilka innych podejść: scatterplot z latami na osi OX/OY, odcinki zaczynające się w danych ze średniej a kończące w 2020. Jestem też ciekawa Waszych innych pomysłów

```python
labels = ['ZACHODNIOPOMORSKIE', 'WIELKOPOLSKIE', 'WARMIŃSKO-MAZURSKIE', 'ŚWIĘTOKRZYSKIE', 'ŚLĄSKIE', 'POMORSKIE',
         'PODLASKIE', 'PODKARPACKIE', 'OPOLSKIE', 'MAZOWIECKIE', 'MAŁOPOLSKIE', 'ŁÓDZKIE', 'LUBUSKIE', 'LUBELSKIE',
         'KUJAWSKO-POMORSKIE', 'DOLNOŚLĄSKIE']
```

```python
srednie = np.array([1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1930, 1768, 2683])
obecnie = np.array([1874, 4013, 1565, 2008, 6270, 2521, 1478, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665])
```

```python
przyrosty = obecnie - srednie
```

```python
plt.scatter(labels, przyrosty)
plt.xticks(labels, rotation='vertical')
plt.ylabel('Przyrost zgonów')
plt.xlabel('Województwo')
plt.show()
```
