```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Zadanko 1
x = np.linspace(-3,3,100)
y = x**3 + 1
plt.plot(x, y)
plt.title("Wykres funkcji f(x) = x^3 + 1", loc = "center")
plt.ylabel('f(x)')
plt.xlabel("x")
```

```python
plt.plot(x, y, label = "f(x)")
plt.title("Wykres funkcji f(x) = x^3 + 1", loc = "center")
plt.ylabel('f(x)')
plt.xlabel("x")

y_sin = np.sin(x)
plt.plot(x, 10*y_sin, color = "purple", linewidth = 3, label=("10*sin(x)"))
plt.ylabel("10*sin(x)")

plt.legend()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, label = "f(x)")
plt.title("Wykres 1", loc = "center")
plt.ylabel('f(x)')
plt.xlabel("x")

plt.subplot(2, 1, 2)
y_sin = np.sin(x)
plt.plot(x, 10*y_sin, color = "purple", linewidth = 3, label=("10*sin(x)"))
plt.ylabel("10*sin(x)")
plt.title("Wykres 2", loc="center")
```

```python
#Zadanko 3
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]

l_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]

l_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]

roznica = [None]*16
for i in range(16):
    roznica[i] = l_2020[i] - l_2015_2019[i]

data = {'Wojewodztwo':wojewodztwo,
       'Roznica':roznica}
df = pd.DataFrame(data)


sort_df = df.sort_values('Roznica')

plt.barh(sort_df['Wojewodztwo'], sort_df['Roznica'], color="purple")
plt.title("Rznica liczby osb zmarych w padzierniku pomidzy redni w latach 2015-2019 a w roku 2020 \n w poszczeglnych wojewdztwach" )

```
