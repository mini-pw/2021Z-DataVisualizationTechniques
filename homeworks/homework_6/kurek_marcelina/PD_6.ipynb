```python
import matplotlib.pyplot as plt
import numpy as np
```

# Zadanie 1

```python
x = np.linspace(-3, 3)

plt.plot(x, x**3+1, linewidth = 1, label = "x^3+1")
plt.title("Wykresy funkcji")
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x, 20*np.cos(2*x)-5, color = 'magenta', linewidth = 2, linestyle = "--", label = "20*cos(2x)")
plt.legend()
```

```python
plt.subplot(121)
plt.plot(x, x**3+1, linewidth = 1, label = "x^3+1")
plt.suptitle("Wykresy funkcji", fontsize =14)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc = 1)

plt.subplot(122)
plt.plot(x, 20*np.cos(2*x)-5, color = 'magenta', linewidth = 2, linestyle = "--", label = "20*cos(2x)")
plt.legend(loc = 1)
plt.xlabel("x")
```

# Zadanie 3

```python
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]

rok_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]

lata_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]

roznica = [0]*np.size(rok_2020)
for i in range(np.size(rok_2020)):
    roznica[i] = rok_2020[i] - lata_2015_2019[i]

fig = plt.figure(figsize = (6.4, 5))
plt.barh(wojewodztwo, lata_2015_2019, color = 'white')
plt.xlabel("Liczba zgonw")
plt.suptitle("Przyrost zgonw w 2020r wzgldem redniej zgonw w latach 2015-2019r", fontsize = 14)

#fig1 = plt.figure()
x = np.linspace(lata_2015_2019[0], rok_2020[0])
s = np.size(x)

for i in range(np.size(wojewodztwo)):
    x = np.linspace(lata_2015_2019[i], rok_2020[i])
    plt.plot(x, [i]*np.size(x), color = 'blue')
    
for i, v in enumerate(roznica):
    plt.text(lata_2015_2019[i] + roznica[i]/3, i + 0.3, str(v), color='black')
    
```
