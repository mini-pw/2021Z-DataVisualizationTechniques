### Zadanie 1.

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

### Rozwiązanie zadania 1a

```python
x = np.linspace(-3, 3, 500)
y = x**3 + 1

plt.plot(x, y)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Wykres funkcji $f(x) = x^3 + 1$")

plt.show()
```

### Rozwiązanie zadania 1b

```python
dy = 3*x**2

plt.plot(x, y, label="$f(x) = x^3 + 1$")
plt.plot(x, dy, color = "green", linewidth = 3, label="$f'(x) = 3x^2 $")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.title("Wykres funkcji $f(x) = x^3 + 1$ wraz z jej pochodną")
plt.legend()

plt.show()
```

### Rozwiązanie zadania 1c

```python
fig, ax = plt.subplots(1, 2, sharey = True)

ax[0].plot(x, y)
ax[0].set_xlabel("$x$")
ax[0].set_ylabel("$y$")

ax[1].plot(x, dy)
ax[1].set_xlabel("$x$")

plt.show()
```

```python
wojewodztwo = np.array(["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMIŃSKO-MAZURSKIE", "ŚWIĘTOKRZYSKIE",  "ŚLĄSKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MAŁOPOLSKIE", "ŁÓDZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOŚLĄSKIE"])
l_2020 = np.array([1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665])
l_2015_2019 = np.array([1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683])

difference = np.subtract(l_2020, l_2015_2019)
sort = sorted(range(len(difference)), key = lambda k: difference[k])
wojewodztwo = wojewodztwo[sort]
l_2020 = l_2020[sort]
l_2015_2019 = l_2015_2019[sort]
```

```python
plt.figure(figsize=(16, 9))
width = 0.8

plt.bar(wojewodztwo, l_2020, width = width, color = "#beaed4", label = "Liczba osób zmarłych w październiku 2020 r.")
plt.bar(wojewodztwo, l_2015_2019, width = width/1.33, color = "#fdc086", label = "Średnia liczba osób zmarłych w październiku w latach 2015–2019")
plt.xticks(rotation = 90)
plt.title("Porównanie liczby osób zmarłych w poszczególnych województwach w październiku 2020 r. i średnio w latach 2015-2019. \n Dane zostały przedstawione od najmniejszej różnicy zgonów do największej.")
plt.legend()

plt.show()
```

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```
