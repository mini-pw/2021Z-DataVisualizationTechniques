# Zadanie1

```python
x=np.linspace(-3,3,1000)
y=pow(x,3)+1
z=-pow(x,2)
plt.plot(x,y,label="x^3+1")
plt.plot(x, z, color = "red", linewidth = 4, label = "-x^2")
plt.title("Wykres",loc="center")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
fig.suptitle('Dwa wykresy')
ax1.plot(x,y,label="x^3+1")
ax2.plot(x,z,color="red",linewidth = 4,label = "-x^2")
ax1.set_title("x^3+1")
ax2.set_title("-x^2")
ax1.set_xlabel("x")
ax2.set_xlabel("x")
ax1.set_ylabel("y")
```
