## Poprawiony wykres z pracy domowej nr 2

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
countries = ['USA', 'Wielka Brytania', 'Niemcy', 'Hiszpania', 'Wlochy', 'Australia', 'Francja', 'Polska', 'Portugalia', 'Rumunia', 'Austria', 'Czechy', 'Holandia', 'Szwajcaria', 'Slowacja', 'Belgia', 'Izrael', 'Szwecja']
total_tests_per_country = [115347287, 26679020, 18129900, 13689776, 12197500, 7935463, 11900095, 3599722, 2764867, 2627544, 1756856, 1515912, 2656731, 1470894, 528367, 3545600, 426328, 1789328]
tests_per_million = [347922, 392435, 216200, 292768, 201821, 310224, 182201, 95483, 271371, 136837, 194765, 141483, 154954, 169617, 96764, 305393, 426328, 176874]
data = pd.DataFrame({'country': countries, 'total_tests': total_tests_per_country, 'tests_per_million': tests_per_million})
data_1 = data.sort_values(by=["tests_per_million"])
data_2 = data.sort_values(by=["total_tests"])
```

```python
fig, ax = plt.subplots(2,1, figsize=(10, 12))
width = 0.8

plt.rcParams['font.size'] = 14

rects2 = ax[0].barh(data_2.loc[:,"country"], data_2.loc[:,"total_tests"], width, color="#007200")
rects1 = ax[1].barh(data_1.loc[:,"country"], data_1.loc[:,"tests_per_million"], width, color="#007200")


def autolabel(rects, sub_plt):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        width = rect.get_width()
        sub_plt.annotate('{:,}'.format(width),
                    xy=(width, rect.get_y() + rect.get_height() / 2),
                    xytext=(3,0),
                    textcoords="offset points",
                    va='center')
       


ax[0].axes.get_xaxis().set_visible(False)
ax[0].set_frame_on(False)

ax[1].axes.get_xaxis().set_visible(False)
ax[1].set_frame_on(False)

autolabel(rects2, ax[0])
autolabel(rects1, ax[1])

ax[0].set_title("Liczba wykonanych testw")
ax[1].set_title("Liczba wykonanych testw na milion mieszkacw")

fig.tight_layout()
plt.show()
```

## Uzasadnienie
1. Kady supek ma teraz przypisan warto, dziki czemu mona atwo odczyta najmniejsze wartoci co byo trudne na wczeniejszych wykresach. 
2. Kady z dwch wykresw jest posortowany odpowiednio wedug liczby wszystkich wykonanych testw lub liczby testw na milion mieskzacw, co uatwia porwnywanie pastw pomidzy sob.
