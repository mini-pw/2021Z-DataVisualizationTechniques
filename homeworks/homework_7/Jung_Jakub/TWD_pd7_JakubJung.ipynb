```python
import matplotlib.pylab as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
import numpy as np
import pandas as pd


testy = pd.read_csv("C://CSV_Files/testy.csv")
testy = testy[testy.Date == "2020-10-11"]
selected_countries = ["USA", "UK", "Germany", "Spain", "Italy", "Australia", "France", "Poland", "Portugal", "Romania", "Austria", "Czechia", "Netherlands", "Switzerland", "Slovakia", "Belgium", "Israel"]
tests = [191037015, 42605903, 27859242, 21917246, 21637641, 9975892, 20418493, 6187205, 4351221, 4097782, 3061067, 3044925, 4043482, 2679348, 1073571, 5871418, 5605071]
df = pd.DataFrame({"Country":selected_countries, "Tests":tests}).sort_values("Tests", ascending=False)
plt.figure(figsize=(22,10))
plt.rc('ytick', labelsize=12) 
plt.rc('xtick', labelsize=12) 
plt.bar("Country", "Tests", data = df)
plt.yscale('log',base=10)
plt.title(label = "Number of tests in different countries", fontsize = 20)
plt.xlabel("Country", fontsize=15)
plt.ylabel("Number of tests", fontsize = 15)

plt.show()

#MÃ³j wykres zrobiony w Pythonie. TworzÄ…c go chciaÅ‚em jednak powrÃ³ciÄ‡ do informacji zawartych na pierwszym wykresie, na ktÃ³rym paÅ„stw byÅ‚o zdecydowanie wiÄ™cej.
#Osoba poprawiajÄ…ca go w ramach PD2 usunÄ™Å‚a czÄ™Å›Ä‡ paÅ„stw, aby zwiÄ™kszyÄ‡ czytelnoÅ›Ä‡ wykresu, ale wykres moÅ¼e pozostaÄ‡ czytelny nawet ze wszystkimi paÅ„stwami.
#Dodatkowo w PD2 usuniÄ™te zostaÅ‚y Stany Zjednoczone, poniewaÅ¼ swojÄ… zdecydowanie wiÄ™kszÄ… liczbÄ… wykonanych testÃ³w zaburzaÅ‚y one wykres.
#Aby tego uniknÄ…Ä‡ wprowadziÅ‚em skalÄ™ logarytmicznÄ…, ktÃ³ra pozwala na przedstawienie wszystkich wynikÃ³w na raz. Wymaga ona jednak duÅ¼o wiÄ™kszego zaangaÅ¼owania
#czytelnika w interpretacjÄ™ przedstawionych danych. Poza tym usunÄ…Å‚em jeszcze kolory, ktÃ³re nie wprowadzaÅ‚y do wykresu Å¼adnej informacji, oraz posortowaÅ‚em
#wyniki dla przyjemniejszego czytania danych.
```
