# Wykres poprawiony

```python

name = [" ","Procesy spalania \n poza przemysłem","Transport drogowy", 
           "Procesy spalania \n w sektorze energii", "Procesy spalania \n w przemyśle", 
           "Inne pojazdy\n i urządzenia", "Procesy\n produkcyjne", "Zagospodarowanie\n odpadów"]
PM10 = [55, 10, 11, 9, 0, 9, 6]
PM2= [55, 12, 11, 8, 8, 6, 0]
X = np.arange(7)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.grid(color='grey', linestyle='--', linewidth=1, alpha =0.4 )
ax.bar(X, PM2, color='b', width =0.3, alpha=0.5, label='Źródła emisji PM 10')
ax.bar(X+0.35, PM10, color="g", alpha =0.5,width =0.3, label='Źródła emisji PM 2.5')
ax.set_xticklabels(name)
ax.set_yticklabels(["0%", "10%", "20%", "30%", "40%", "50%"])
plt.xticks(rotation=45)
plt.legend()
plt.title("Procentowe źródła emisji PM w 2016r. w Polsce",size=18)
plt.show()
```
