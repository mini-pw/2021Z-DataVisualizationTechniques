## Drugi poprawiony wykres

```python
dates = ["11/06", "30/06", "01/07", "31/07", "05/08", "31/08", "04/09", "26/09", "01/10", "08/10", "15/10"]
x = [dt.datetime.strptime(d,'%d/%m').date() for d in dates]
y = [75, 85, 78, 72, 72, 83, 75, 110, 159, 296, 508]


fig, ax = plt.subplots(figsize = (12,9))

rects = ax.bar(x,y, color = "#FFFFFF")

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d/%m'))
plt.xlabel('wybrane dni od czerwca do padziernika', size = 15)
plt.title("Liczba zajtych respiratorw w cigu ostatnich miesicy", size = 23)
plt.grid(b=True, which='major', axis='y')
plt.rc('axes', axisbelow=False)

for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., height+10,
                '%d' % int(height),
                ha='center', va='bottom')

plt.plot(x, y, 'o-', color = "#956D1B")
        
plt.show()
```

Wykres pokazuje nieregularno w przedstawionych danych oraz przez zmian rodzaju wykresu zaznacza, e pokazana warto zmienia si w czasie.
