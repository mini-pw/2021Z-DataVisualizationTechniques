## Wykres poprawiony 2

```python
import matplotlib.pyplot as plt
import numpy as np
import datetime

fig=plt.figure(figsize=[10,5])
plt.grid(color="gray", axis="y", linestyle=':', linewidth=0.5)
resp=[159,166,179,191,219,263,283,296,320,346,383,404,421,467,508]
start = datetime.datetime.strptime("01-10", "%d-%m")
end = datetime.datetime.strptime("16-10", "%d-%m")
date = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]
for i in range(len(date)):
    date[i]=date[i].strftime("%d-%m")

x1=np.arange(15)
plt.bar(date,resp,color="steelblue")

for i in range(len(date)): 
    plt.text(x1[i-1]-0.33,resp[i-1]+5,resp[i-1],fontsize=12,color="black")
plt.ylim(0,550)
plt.title("Liczba zajetych respiratorow")

```
