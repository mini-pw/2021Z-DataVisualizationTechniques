# Poprawiona wersja

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
sns.set(rc={'figure.figsize':(12,7)})
sns.set_style('whitegrid')
ax = sns.barplot(data=df, x = 'location', y = 'total_tests')

for bar in ax.patches:
 height = bar.get_height()
 ax.text(x = bar.get_x()+(bar.get_width()/2), y = height+1000000, s = "{:,}".format(round(height)), ha = 'center')

ax.set_ylabel("Liczba przeprowadzonych testw [10^8]", fontsize = 14)
ax.set_xlabel(None)
ax.set_title("Suma przeprowadzonych testw do 11 padziernika 2020", fontsize = 16)
plt.show()
```
