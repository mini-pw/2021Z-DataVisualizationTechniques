```python
import matplotlib.pyplot as plt
```

```python
data = [66,20,7,6]
dataone = [data[0] ,sum(data[1:])]
labsone = ["Kartą","Pozostałe"]
labsone = [el + "\n" + str(d)+"%" for el,d in zip(labsone,dataone)]
datatwo = data[1:]
datatwo = [100*el/(sum(datatwo)) for el in datatwo]
labstwo = ["Gotówką","Telefonem/zegarkiem","Bankowością internetową"]
labstwo = [el + "\n" + str(d)[:2]+"%" for el,d in zip(labstwo,datatwo)]

sizeone = 1.3
f,a = plt.subplots(2,1)
a[0].pie(dataone,radius=sizeone,labels=labsone,colors=["#323031","#ffc857"],startangle=60)
a[0].add_artist(plt.Circle((0,0),0.6*sizeone,color="white"))
f.suptitle("Czym najczęściej płacą polacy?",weight="bold")

size = 0.66 * sizeone
a[1].pie(datatwo,radius=size,labels=labstwo,colors=["#117e89","#084c61","#DB3A34",],startangle=60,labeldistance=1.3)
a[1].add_artist(plt.Circle((0,0),0.6*size,color="white"))
a[1].set_title("Podział pozostałych",loc="right")
plt.show()
```

# Czemu wykres w pythonie jest lepszy?
Celem wykresu jest pokazanie tezy, że polacy najczęściej płacą kartą, co jest bardzo wyraźnie czytelne na pierwszym donacie. Odbiorca potrzebujący dodatkowych informacji może je z łatwością odczytać z drugiego wykresu. Przez rodzdzielenie danych na dwa wykresy, od razu bez czytania wartości procentowych widać fakt, że płatność kartą dominuje wśród polaków, a na drugim wykresie widać od razu że płatność gotówką dominuje wśród pozostałych metod płatności.

Ponadto, użycie kołowopochodnego wykresu nie wpływa negatywnie na łatwość odczytywania danych - wykres ten nie ma na celu pokazanie dokładnych procentowych wartości w danych, tylko procentowy udział poszczególnych kategorii, co pokazuje czytelnie.
