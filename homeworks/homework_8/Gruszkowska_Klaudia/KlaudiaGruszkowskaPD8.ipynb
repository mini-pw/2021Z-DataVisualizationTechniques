## Wykres poprawiony

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#DANE
ilosc = [12,100,20,20,48,21,42,107,35]
labels = ["tundra areas","boreal forest","temperate grasslands","temperate coniferous woodlands","temperate broadleaf woodlands","tropical dry areas","tropical grasslands","tropical rainforests","other"]
```

```python
#WYKRES
plt.figure(figsize =(10,10))
ilosc = pd.Series([12,100,20,20,48,21,42,107,35])
labels = pd.Series(["tundra areas","boreal forest","temperate grasslands","temperate coniferous woodlands","temperate broadleaf woodlands","tropical dry areas","tropical grasslands","tropical rainforests","other"])
df = pd.concat([ilosc,labels],axis=1,join="inner")
df.columns = ["ilo","miejsce"]
df = df.sort_values("ilo", ascending=True)
df[["ilo"]]
plt.barh(df["miejsce"],df["ilo"],color = "#009933")
plt.title("For each person on the planet there are 400 trees",fontsize = 20)
plt.xlabel("number of trees",fontsize =15)
plt.grid(axis='x', linestyle='--', linewidth=1)
        
plt.show()
```
