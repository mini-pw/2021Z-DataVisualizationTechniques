## Poprawiona wizualizacja

```python
# bilbioteki
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.ticker import FuncFormatter



barWidth = 0.5
 
# dane
nieKobiety = [3.3, 15.7, 9.0, 14.3, 15.5, 17.9, 19.2]
nieMezczyzni = [4.8, 9.3, 13.4, 16.6, 10.4, 12.3, 15.7]

nieWiemKobiety = [19.9, 32.4, 31.4, 26.3, 35.6, 34.7, 39.3]
nieWiemMezczyzni = [28.0, 33.8, 37.2, 34.3, 34.1, 50.3, 46.9]

takKobiety = [76.8, 51.9, 59.5, 59.4, 48.9, 47.4, 41.6]
takMezczyzni = [67.3, 56.9, 49.4, 49.1, 55.5, 37.4, 37.3]


# pozycja slupkow na osi
r1 = np.arange(len(nieKobiety))
r1 = [x + x*barWidth for x in r1]
r2 = [x + barWidth for x in r1]



# Rysujemy

plt.figure(figsize=(20,7))
plt.rc('axes', axisbelow=True)
plt.grid(axis = "y")



plt.bar(r1, nieKobiety, color='brown', width=barWidth, edgecolor='white', label='var1', hatch = "/")
plt.bar(r2, nieMezczyzni, color='brown', width=barWidth, edgecolor='white', label='var1')

plt.bar(r1, nieWiemKobiety, bottom = nieKobiety, color='gold', width=barWidth, edgecolor='white', label='var2', hatch = "/")
plt.bar(r2, nieWiemMezczyzni, bottom = nieMezczyzni, color='gold', width=barWidth, edgecolor='white', label='var2')

plt.bar(r1, takKobiety, bottom = [i+j for i,j in zip(nieKobiety, nieWiemKobiety)], color='darkgreen', width=barWidth, edgecolor='white', label='var3', hatch = "/")
plt.bar(r2, takMezczyzni, bottom = [i+j for i,j in zip(nieMezczyzni, nieWiemMezczyzni)], color='darkgreen', width=barWidth, edgecolor='white', label='var3')
 



#legenda
colors = ['darkgreen','gold','brown']

a_val = 1
lab1 = mpatches.Patch(facecolor="white", alpha= 0, label="Odpowiedź")
lab2 = mpatches.Patch( facecolor=colors[0],alpha=a_val,label='Tak')
lab3 = mpatches.Patch( facecolor=colors[1],alpha=a_val,label='Nie wiem')
lab4 = mpatches.Patch(facecolor=colors[2],alpha=a_val,label='Nie')
przerwa = mpatches.Patch(facecolor="white", alpha= 0, label="")
lab5 = mpatches.Patch(facecolor="white", alpha= 0, label="Płeć")
lab6 = mpatches.Patch(facecolor="white",alpha=a_val,edgecolor="black",hatch='/////',label='Kobieta')
lab7 = mpatches.Patch(facecolor="white",alpha=a_val,edgecolor="black",label='Mężczyzna')


plt.legend(handles = [lab1, lab2, lab3, lab4, przerwa, lab5, lab6, lab7],loc=1, bbox_to_anchor=(1.09, 1))

#tytul i estetyka
plt.title('W jakim stopniu zgadzasz się ze stwierdzeniem...',
         fontsize = 34)

etykiety = ['Bardzo cenię sobie\n niezależność i podążanie\n własną ścieżką zawodową', 'Kobiety i mężczyźni\nróżnią się predyspozycjami\ni dlatego różnie sprawdzają\nsię w różnych zawodach', 'Uważam, że kobiety\nsprawdzają się w STEM czyli\nw zawodach i naukach\nścisłych, technologii inżynierii\ni matematyce tak samo\ndobrze jak mężczyźni', 'Kobiecie jest trudniej\nniż mężczyźnie budować\nkarierę zawodową\nze względu\nna zakładane rodziny', 'Jest dla mnie bardzo\nważne, aby spełniać\noczekiwania moich bliskich','Kobiety wybierają karierę,\nktóra mniej ingeruje\nw ich życie prywatne','Pociągają mnie zajęcia\nniszowe, niepopularne']
plt.xticks([(i+j)/2 for i,j in zip(r1, r2)], etykiety)


plt.box(False)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.0%}'.format(y/100))) 


plt.show()
```
