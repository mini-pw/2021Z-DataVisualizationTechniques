```python
Months = ['11.2019', '12.2019', '01.2020', '02.2020', '03.2020', '04.2020', '05.2020', '06.2020', '07.2020', '08.2020', '09.2020',
         '10.2020', '11.2020']
Values = [2.6, 3.4, 4.3, 4.7, 4.6, 3.4, 2.9, 3.3, 3.0, 2.9, 3.2, 3.1, 3.0]
Months.reverse()
Values.reverse()

plt.figure(figsize=(12, 8))
plt.barh(Months, Values)
plt.xlabel("Zmiana w % w stosunku do tego samego miesica poprzedniego roku", fontdict = {'fontsize': 15})
plt.title("Ceny towarw i usug konsumpcyjnych", 
          fontdict = {'fontsize': 20})
plt.ylabel("Miesic", fontdict = {'fontsize': 15})
plt.grid(axis = 'x', lw = 0.4, c = 'gray')
plt.xticks(np.linspace(0, 5, 11))
for i in range(0, len(Months)):
    plt.text(Values[i] + 0.04, Months[i] , Values[i], fontsize = 13)
plt.show()
```
