## Wykres poprawiony

```python
import matplotlib.pyplot as plt
import numpy as np


fig=plt.figure(figsize=[10,5])


plt.grid(color="gray", axis="y", linestyle=':', linewidth=0.5)
y=[-0.3,1.9,0.6,-0.3,-8.4]
x=[1/40,2/40,3/40,4/40,5/40]
y1=[0,0]
x1=[0,1.4]
plt.plot(x1,y1,color="lightgray")
text=["All-items HICP", "Food, alcochol & tobacco","Energy","Non-energyindustrial goods","Services"]

plt.bar(x,y,color="steelblue",width=0.003)
plt.bar(1/40,-0.3,color="red",width=0.003)

for i in range(len(x)):
    if(y[i]<0):
        y[i]-=0.6
    else:
        y[i]+=0.4
    plt.text(x[i]-len(text[i])/1280,y[i],text[i],fontsize=12,color="black")
plt.ylim(-10,3)
plt.xlim(0,0.14)
plt.title("Euro area annual inflation, November 2020, %",fontsize=20)
plt.xticks([])
```

# Uzasadnienie
Zamieniem orientacj wykresy na pionow, aby uatwi porwnanie wartoci dodatnich i ujemnych.Dodatkowo dziki temu zabiegowi podpisyu supkw nie s poprzecinane liniami pomocniczymi wykresu. Zmieniem styl lini pomocniczych, na taki, ktry bardziej wtapia si w to, przez co jest mniej rozpraszajcy.
