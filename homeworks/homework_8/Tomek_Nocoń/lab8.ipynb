```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
from IPython.display import Image
Image(filename="Chart_05_Percentage_employment_income_loss_by_gender_age_group_and_transition_type.png")
```

```python
incomeLossMale = pd.DataFrame(np.array([[5,"16-24","to absent from work"], [1, "16-24","to working reduced hours(80% or less)"], 
                                    [5.2 , "16-24","to unemployed"], [3.3,"25-65", "to absent from work"],
                                   [1.1, "25-65", "to working reduced hours(80% or less)"], [0.5, "25-65","to unemployed"]]),
                         columns=['percent', 'age', 'state'])
incomeLossFemale = pd.DataFrame(np.array([[5.1,"16-24","to absent from work"], [1.1, "16-24","to working reduced hours(80% or less)"], 
                                    [6 , "16-24","to unemployed"], [3,"25-65", "to absent from work"],
                                   [1, "25-65", "to working reduced hours(80% or less)"], [0.5, "25-65","to unemployed"]]),
                         columns=['percent', 'age', 'state'])
```

```python


sns.catplot(x="age", y="percent", hue="state", kind="bar", data=incomeLossMale)
plt.title('Percentage employment income loss among male, by age group and transition type, EU-27, %')
sns.catplot(x="age", y="percent", hue="state", kind="bar", data=incomeLossFemale)
plt.title('Percentage employment income loss among female, by age group and transition type, EU-27, %')
plt.show()
```
