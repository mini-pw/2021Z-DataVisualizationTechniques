```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 10)
y_sin = np.sin(x)
print(x)
```

# Wykres liniowy

```python
plt.plot(x, y_sin)
plt.show() #- potrzebne w niekt√≥rych IDE np terminal lub skrypty.
```

## modyfikacje

# Figury

`plt.plot()` pozwala na narysowanie tylko jednej figury o wsp√≥≈Çrzƒôdnych (x,y). Je≈õli chcemy jednƒÖ grafikƒô, mo≈ºna przy tym zostaƒá.

```python
names = ['A', 'B', 'C']
values = [1, 10, 25]
plt.bar(names, values)
plt.show()
```

`plt.figure()` definiujemy figurƒô, a kt√≥rej mo≈ºemy potem umieszczaƒá inne elementy).

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1) # pozycja wykresu na figurze: nrows, ncols, index
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.bar(names, values)
plt.show()
```

`plt.subplot()` Pozwala na narysowanie jednej lub kilku figur w tym samym oknie. 

Poni≈ºej rysujemy dwie figury `ax1` i `ax2` w tym samym oknie, jedynie podzielonym na dwie czƒô≈õci.

```python
fig, (ax1, ax2) = plt.subplots(2, 1) # 2 wiersze, 1 kolumna
fig.suptitle('Subplot')

ax1.bar(names, values)
ax2.bar(names, [1,1,2])

ax1.set_ylabel("value")
ax2.set_ylabel("value")

plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

ax[0,0].bar(names, values)
ax[1,1].bar(names, [1,1,2])
ax[1,0].barh(names, [1,1,2])

plt.show()
```
